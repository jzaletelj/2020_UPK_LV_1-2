## Matplotlib : risanje grafov

Matplotlib je najpomembnejša knjižnjica za risanje grafov v Pythonu, zgleduje se po Matlabu.

Spletna stran: http://matplotlib.org/

    
### Uvoz knjižnjice

Uvozimo `matplotlib.pyplot` modul pod imenom `plt`:

```python
import matplotlib.pyplot as plt
```

Potrebujemo še naslednje za uporabo v Jupyter beležnici.

```python
%matplotlib inline
```

### Primer

Ustvarimo dve polji podatkov, ki jih izrišemo.

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

## Osnovni ukazi risanja grafa

Osnovna metoda za risanje grafa je plot

```python
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X os : stevila')
plt.ylabel('Y : kvadrati stevil')
plt.title('Naslov grafa')
plt.show()
```

### Naloga: 
Spremeni y funkcijo in  ponovno izriši graf. Preveri opcije metode plot(). Preveri, kako lahko vplivamo na barvo in način izrisa črte.

## Več grafov na istem polju

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') 
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

## Posebni tipi grafov



```python
plt.scatter(x,y)
```

```python
from random import sample
data = sample(range(1, 1000), 100)
plt.hist(data)
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);   
```
